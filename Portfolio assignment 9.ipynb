```python
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins.head()
```

```python
penguinsCorrelations = penguins.corr()
penguinsCorrelations.style.background_gradient(cmap='coolwarm', axis=None).format(precision=2)
```

```python
# Yes, they match my expectations of correlations, because the more length the more mass, 
# bill length and depth have a difference around 20, and bill depth and wing length have a difference around 160.
```

```python
penguins.plot(kind='scatter', x='bill_length_mm', y='bill_depth_mm') # Weakest
```

```python
penguins[['bill_length_mm','bill_depth_mm']].corr()
```

```python
penguins.plot(kind='scatter', x='bill_depth_mm', y='flipper_length_mm') # Strongest negative
```

```python
penguins[['bill_length_mm','flipper_length_mm']].corr()
```

```python
penguins.plot(kind='scatter', x='flipper_length_mm', y='body_mass_g') # Strongest positive
```

```python
penguins[['flipper_length_mm','body_mass_g']].corr()
```
